## KNN and some issue about ML 

***semi-supervised learning**: một thuật toán kết hợp giữa học có giám sát và học không giám sát. Data ban đầu sẽ chỉ được label một phần nhỏ từ trước, sau đó dùng supervised learning với phần data còn lại chưa được gắn nhãn. kết quả là một data được supervised hoàn toàn*  

***Eager-learning apporoach (thích học)**: dữ liệu được huấn luyện trước. khi cần giải quyết các sample, thuật toán sẽ chạy nhanh hơn.*  

***lazy learning approach**: dữ liệu chỉ được huấn luyện khi thực hiện bài toán. mất nhiều thời gian để xử lý hơn vì model sẽ không học từ đầu.*  

***distance**: thực chất là similarity feature (độ tương dồng của 2 điểm dữ liệu)*

```python
# lay hang cuoi cung de thuc hien kiem thu bang KNN, cac hang tren duoc dung de train

testingSet = data[-1]
testingSet.shape
```

### Eculid Distance

```python
def euclide_distance(data_point01, data_point02):
    result = 0

    for i in range(4):
        result += (float(data_point01[i]) - float(data_point02[i]))**2
        
    return np.sqrt(result)
```

### compute K Nearest Neighbor. select K such that arrcordingly.

```python
def compute_knn(arr, k):    
    # sort khong can goi tham so
    arr.sort(key=lambda x: x['distance'])

    arr = arr[:k]
    return [item['type'] for item in arr]
```

### voting

```python
def vote_type(arr):
    maximize = 0 
    type_flowers = ''
    union = set(arr)
    for element in union:
        if (arr.count(element) > maximize):
            maximize = arr.count(element)
            type_flowers = element

    return type_flowers
```

## main func

```python
if __name__ == "__main__":
    distance = []
    for i in range(len(trainingSet)):
        distance.append(
            {
            'distance': euclide_distance(testingSet, trainingSet[i]),
            'type': trainingSet[i][4]
            }
        )

    distance = compute_knn(distance, 9)
    result = vote_type(distance)
    print(result)
```
